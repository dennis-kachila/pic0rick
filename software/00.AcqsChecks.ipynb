```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("acqs2.md","r") as f:
    T = f.read()
```

```python
F = [x.replace("\n","") for x in T.split("----------") if len(x.replace("\n","")) > 300]
t = F[1]
```

```python
[len(x) for x in F]
```

```python
s = [int(x,16) for x in t.split(",") if len(x)]
ti = [x/20 for x in range(len(s))]
```

```python
s = [int(x,16) for x in t.split(",") if len(x)]
ti = [x/60 for x in range(len(s))]
plt.figure(figsize=(15,5))
plt.xlabel("micro seconds")
plt.ylabel("ADC value")
plt.title("Pulse echo !")
plt.plot(ti,s)
```
